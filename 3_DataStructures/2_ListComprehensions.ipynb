```python
##List Comprehensions

print('''
List Comprehension is unique way of quickly creating a list with python.
if we are using for loop along with append() to create a list.
List Comprehension are a good alternative.

syntax :

list=[operationToPerform for in variable]

list comprehension uses greedy evaluation - they create list immediately.
''')
```

```python
mystring="Hello World"
## old way
mylist=[]
for l in mystring:
    mylist.append(l)
print(mylist)

## Other efficient way using  List Comprehensions
list=[l for l in mystring]
print(list)
```

```python
mylist=[a for a in "Python"]
print(mylist)
```

```python
num_list=[num for num in range(5,50,5)]
print(num_list)
```

```python
## Get Square
num_lst=[num**2 for num in range(5,50,5)]
print(num_lst)
```

```python
## Get Even number using if loop
even_list=[num for num in range(20) if num%2==0]
print(even_list)
```

```python
## convert temperature Celsius to Fahrenheit

celcius=[10,34,67,34]
farhenit_list=[((9/5)*temp+32) for temp in celcius]
print(farhenit_list)
```

```python
## if else loop - not recommended
results = [x if x%2==0 else 'ODD' for x in range(10)]
print(results)
```

```python
## nested for loop

mylist=[x*y for x in range(10) for y in range(10) ]
print(mylist)
```

```python
mylist=[l if l in ['a','e','i','o','u'] else '0' for l in 'aasavfvgffrifhvbdoivvmc']
print(mylist)
```

```python
items=["apple","banana","orange","mango"]
upper_items=[item.upper() for item in items]
print(upper_items)
```
