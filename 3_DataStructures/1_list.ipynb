```python

list=[1,4,5,True,"Test"]

## list iteration
for value in list:
    print(value)

## List methods
print(f'the length of list is {len(list)}')
print(list.count(5))  #(count occurrences)
print(list.index(5))   #(find position)

#sort
#list.sort() -> Won't work if list contains other than number
num_list=[2,5,3,7,1,4,12,15,11,10]
print(num_list)
num_list.sort()
print(num_list)

#Reverse
num_list.sort(reverse=True)
print(num_list)

##one more reverse()
num_list.reverse()
print(num_list)

##remove
num_list.remove(2)
print(num_list)
```

```python
## Copying the list -copy()
list=[3,5,6,7,1,2,45]
print(list)
new_list=list.copy()
print(new_list)
new_list.remove(45)
print(new_list)
print(list)
```

```python
## Common Mistakes

# Index out of range
list=[8,9]
#print(list[2]) #-IndexError: list index out of range

# Modifying while looping
numbers = [1, 2, 3, 4]
for num in numbers:
    if num==2:
        numbers.remove(num) # Dangerous!
# Right - use list comprehension
numbers = [num for num in numbers if num != 2]

# Shallow copy issues
# Wrong - both variables point to same list
list1 = [1, 2, 3]
list2 = list1
list2.append(4)
print(list1)  # [1, 2, 3, 4] - changed!

# Right - make a copy
list1 = [1, 2, 3]
list2 = list1.copy()
list2.append(4)
print(list1)  # [1, 2, 3] - unchanged
```

```python
##zip ->combines list and returns tuple

myList1=[1,2.3,5]
myList2=['a','b','c']

zip(myList1,myList2)

for item in zip(myList1,myList2):
    print(item)
```
