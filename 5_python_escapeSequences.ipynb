```python
## \n - > new line
print("Hello\nWorld")
```

```python
## \t - > insert horizontal tab
print("Hello\tWorld to my\t workspace")
```

```python
## \\ insert backslash character
print("Hello\\World")
```

```python
## \" -> insert double quote
print("Hello\"World")
```

```python
## \' Insert single quote
print("Hello\'World")
```

```python
## / ignoring line break in string
print("Hello This is Anand and we have written \
We have spllited these sentence")
```
