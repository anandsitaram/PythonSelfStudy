```python
##  Numeric Data type
a=10   #int
b=78.89  #float
c=8+7j   #complex
print("Data Type of a ",type(a))
print("Data Type of b ",type(b))
print("Data Type of c ",type(c))
```

```python
## Boolean Data Type
val=False
print(val)
print(type(val))
print(val==False)
```

```python
## string Data type
str="Double Quote String"
val='Single Quote String'
conc=str+val
a=90
#concc=str+a - This won't work we can't concatenate string with non string data type
print("Data type of str ",type(str))
print("Data type of val ",type(val))
print("Data type of conc ",type(conc))
```

```python
## List Data Type
values = [1,5,8.9,True,"Faliu"]
print(values)  #print all list
print(values[0])# print first value
print(values[3])# print third value
print(values[-1]) #last value
print(len(values)) #length of value [list]
print(values[0:4])  #sub list with last value is exclusive
print(values[2:5]) #sub list with last value is exclusive
values.insert(1,"NewValue") #insert "NewValue" text at 1st index
print(values)
values.append("AtEnd") # append at last
print(values)
values[1]="replaceValue"  #update value
print(values)
del values[0] #delete values[0]
print(values)
```
