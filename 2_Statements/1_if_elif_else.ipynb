```python
## can have multiple elif statements and else is optional
color="Black"
if color=="Red":
    print("If condition is true")
elif color=="Blue":
    print("If condition is true")
elif color=="Black":
    print("elif condition is true")
#else is optional
else:
    print("If condition is false")

```
