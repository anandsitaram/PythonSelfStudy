```python
#print only even numbers from lst
lst=[3,6,8,12,14,15,17,19,20]
for num in lst:
    if num%2==0:
        print(f'{num} is even')
    else:
        print(f'{num} is odd')
```

```python
## example for tuple
tup=(4,6,7,8,9)
for num in tup:
    print(num)
```

```python
tup_lst=[(1,3),(5,7),(8,9),(12,15)]
for num in tup_lst:
    print(num)

print(f'length of tuple list is {len(tup)}')
for a,b in tup_lst:
    print(a,b)
```

```python
## example of dictionary
dic={"a1":1,"a2":2,"a3":3}

## print only key
for d in dic:
    print(d)

## print key ,value
for key,value in dic.items():
    print(key,value)

## print only values
for value in dic.values():
    print(value)
```

```python
## enumerate -> get index and value

word="python"
for letter in enumerate(word):
    print(letter)
```
