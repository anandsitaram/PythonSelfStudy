```python
## Syntax

'''
while "some_condition_true":
    do_some_thing()
'''
```

```python
x=0
while x<5:
    print(f'The current value of x is {x}')
    x+=1
```

```python
## We can also add else condition in while loop
x=50
while x<5:
    print(x)
    x+=1
else:
    print("x greater than 5")
```

```python
## break - Breaks come out of current closest enclosing loop
x=5
while x<12:
    print(f'The current value of x is {x}')
    if x==8:
        print("value is 8 so breaking the loop")
        break
    x+=1
    print("value is increased")
print("outside while loop")
```

```python
## continue ->goes to top of the closest loop
x=5
while x<12:
    print(f'The current value of x is {x}')
    x+=1
    if x==8:
        print("value is 8 so breaking the loop")
        continue
    print("value is increased")
print("outside while loop")
```

```python
## pass-do nothing
x=5
while x<12:
    print(f'The current value of x is {x}')
    x+=1
    if x==8:
        print("value is 8 so breaking the loop")
        pass
    print("value is increased")
print("outside while loop")
```
