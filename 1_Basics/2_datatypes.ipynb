```python
##  Numeric Data type
a=10   #int
b=78.89  #float
c=8+7j   #complex
print("Data Type of a ",type(a))
print("Data Type of b ",type(b))
print("Data Type of c ",type(c))
```

```python
## Boolean Data Type
val=False
print(val)
print(type(val))
print(val==False)
```

```python
## string Data type
str="Double Quote String"
val='Single Quote String'
conc=str+val
a=90
#concc=str+a - This won't work we can't concatenate string with non string data type
print("Data type of str ",type(str))
print("Data type of val ",type(val))
print("Data type of conc ",type(conc))
```

```python
## List Data Type
#List is mutable - >we can change\update the value in list
values = [1,5,8.9,True,"Faliu"]
print(values) #print all list
print(type(values))
print(values[0])# print first value
print(values[3])# print third value
print(values[-1]) #last value
print(len(values)) #length of value [list]
print(values[0:4])  #sub list with last value is exclusive
print(values[2:5]) #sub list with last value is exclusive
values.insert(1,"NewValue") #insert "NewValue" text at 1st index
print(values)
values.append("AtEnd") # append at last
print(values)
values[1]="replaceValue"  #update value
print(values)
del values[0] #delete values[0]
print(values)
```

```python
## Tuple data type
# tuple is immutable - > we can't change or update the values
tup=(9,7.9,False,'text')
print(tup)
print(type(tup))
print(tup[0]) #first value
print(tup[-1]) #last value
print(len(tup)) #length of tup
print(tup[0:2])  #sub tuple with last value is exclusive
print(tup[2:4]) #sub tuple with last value is exclusive
# tup[0]=10 - >This won't work
```

```python
## Dictionary Data Type
dic={"a":1,"b":'Test',6:False}
print(dic)
print(len(dic))
print(dic["a"])
print(dic["b"])
# print(dic["c"])  - >Invalid key error
print(dic[6])
```

```python
## Sets -> unordered collection of unique elements -No Duplicates
mySet= set() #-> Creating empty set
mySet.add(1)
mySet.add("Test")
print(mySet) # values won't be printing in order
print(type(mySet))
mySet.add(1) # value won't be added
print(mySet)
list=[1,1,1,1,4,5,6,4,5,6,'Test','Test','A',7,5,3,2]
st=set(list)
print(st) #prints unique elements
```
