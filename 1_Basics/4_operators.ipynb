```python
## Arithmetic
a=10
b=3
print("addition is",a+b)
print("multiplication is",a*b)
print("division is",a/b)
print("subtraction is",a-b)
print("modulo is",a%b)
print("exponent is",a**b)
print("floor is",a//b)
```

```python
## Grouping

op=10*(9+10)
print(op)

# without grouping
op=10*9+10
print(op)
print("preference is multiplication,division,modulo and addition,subtraction,exponent")
op=10*9+10
```

```python
## Comparison operators

print(2==2)
print(2=='2')
print("hello"=="Hello")
print(9!=2)
print(5!=5)
print(2>1)
print(1<2)
print(3>=1)
print(4<=1)

print(1<2<5)
```

```python
## Logical Operators
## and
## or
## not

print(5<2 and 3<5)
print(1<2 or 3<5)
print(not(1==1))
print(2 not in [1,2,3,4,5])
```
